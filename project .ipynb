#        sklearn

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model as lm
from sklearn.metrics import mean_squared_error,r2_score 
```

```python
cal_train,cal_test,rat_train,rat_test=train_test_split(cal,rat,test_size=0.3,random_state=1)
sod_train,sod_test,rat_train,rat_test=train_test_split(sod,rat,test_size=0.3,random_state=1)
fib_train,fib_test,rat_train,rat_test=train_test_split(fib,rat,test_size=0.3,random_state=1)
typ_train,typ_test,rat_train,rat_test=train_test_split(typ,rat,test_size=0.3,random_state=1)
pro_train,pro_test,rat_train,rat_test=train_test_split(pro,rat,test_size=0.3,random_state=1)
fat_train,fat_test,rat_train,rat_test=train_test_split(fat,rat,test_size=0.3,random_state=1)
vit_train,vit_test,rat_train,rat_test=train_test_split(vit,rat,test_size=0.3,random_state=1)
she_train,she_test,rat_train,rat_test=train_test_split(she,rat,test_size=0.3,random_state=1)
wei_train,wei_test,rat_train,rat_test=train_test_split(wei,rat,test_size=0.3,random_state=1)
cup_train,cup_test,rat_train,rat_test=train_test_split(cup,rat,test_size=0.3,random_state=1)
sug_train,sug_test,rat_train,rat_test=train_test_split(sug,rat,test_size=0.3,random_state=1)
pot_train,pot_test,rat_train,rat_test=train_test_split(pot,rat,test_size=0.3,random_state=1)
car_train,car_test,rat_train,rat_test=train_test_split(car,rat,test_size=0.3,random_state=1)
```

# calories

# fitting the model

```python
cal_train=cal_train.values.reshape(-1,1)
cal_test=cal_test.values.reshape(-1,1)
```

```python
model1=lm.LinearRegression()
results1=model1.fit(cal_train,rat_train)
results1
```

# finding accuracy of the model

```python
accuracy=model1.score(cal_train,rat_train)
print("accuracy of calories and rating model:",accuracy)
```

```python
predictions=model1.predict(cal_test)
print("predcitions of ratings",predictions,sep='\n')
```

```python
plt.scatter(rat_test,predictions)
```

```python
rat_train_prediction= model1.predict(cal_train)
rat_test_prediction=model1.predict(cal_test)
```

```python
rmse_training=np.sqrt(mean_squared_error(rat_train,rat_train_prediction))
r2_training=r2_score(rat_train,rat_train_prediction)
print('RMSE is {}'.format(rmse_training))
print('R2 is{}'.format(r2_training))
```
